Q2: List the school name, community name, and safety score for all schools with a safety score above 80. Sort by highest safety score.
answer : no one 

```python
q = '''SELECT school.name_of_school, census.community_name, school_report.safety_score
FROM school
JOIN census ON school.`school_id-community_id` = census.community_id
JOIN school_report ON school.`school_id-community_id` = school_report.school_id
WHERE school_report.safety_score > 80
ORDER BY school_report.safety_score DESC;



'''
pd.read_sql(q, conn)
```

Q3: What is the average safety score of all schools when grouped by level of school (ES, MS, HS)? Include the level of school in the result. Order alphabetically by school level. answer : 0 ES 49.520384 1 HS 49.623529 2 MS 48.000000

```python
q = '''SELECT school.school_level, AVG(school_report.safety_score) AS average_safety_score
FROM school
JOIN school_report ON school.`school_id-community_id` = school_report.school_id
GROUP BY school.school_level
ORDER BY school.school_level ASC;


'''
pd.read_sql(q, conn)
```
